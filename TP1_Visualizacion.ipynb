```python
import pandas as pd

# Cargar el archivo CSV en un DataFrame
csv_file = 'C:\\Users\\a119234\\Desktop\\federico.csv'
df = pd.read_csv(csv_file)

# Mostrar las primeras filas del DataFrame
df.head()
```

```python
# Asegurarse de que 'End Date' esté en formato datetime
df['End Date'] = pd.to_datetime(df['End Date'], errors='coerce')

# Ordenar el DataFrame por 'End Date'
df_sorted = df.sort_values(by='End Date')

# Mostrar las primeras filas del DataFrame ordenado
df_sorted.head()
```

```python
# Asegurarse de que 'Start Date' esté en formato datetime
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')

# Ordenar el DataFrame por 'Start Date' en orden descendente
df_sorted_start_desc = df.sort_values(by='Start Date', ascending=False)

# Mostrar las primeras filas del DataFrame ordenado
df_sorted_start_desc.head(50)
```

```python
# Información básica sobre el DataFrame
df.info()

# Estadísticas básicas del DataFrame
df.describe()

# Mostrar nombres de columnas
df.columns
```

```python
# Revisar si hay valores nulos
df.isnull().sum()
```

```python
import pandas as pd
from datetime import datetime

# Definir la fecha límite como el 30 de septiembre de 2024
end_date = datetime(2024, 9, 30)

# Calcular la fecha de inicio (un año antes del 30 de septiembre de 2024)
start_date = end_date - pd.DateOffset(years=2)

# Filtrar las reuniones entre la fecha de inicio y la fecha de fin
df_filtered = df[(df['Start Date'] >= start_date) & (df['Start Date'] <= end_date)]

# Mostrar algunas filas para verificar el resultado
df_filtered.head()
```

```python
# Comprobar las fechas mínima y máxima en el DataFrame filtrado
print("Fecha mínima en 'df_filtered':", df_filtered['Start Date'].min())
print("Fecha máxima en 'df_filtered':", df_filtered['Start Date'].max())
```

```python
# Eliminar las columnas 'Location' y 'Description' del DataFrame
df_filtered = df_filtered.drop(columns=['Location', 'Description'])

# Mostrar algunas filas para verificar el resultado
df_filtered.head()
```

```python
# Crear una nueva columna 'Duration' que sea la diferencia entre 'End Date' y 'Start Date'
df_filtered['Duration'] = (df_filtered['End Date'] - df_filtered['Start Date']).dt.total_seconds() / 60  # Duración en minutos

# Mostrar algunas filas para verificar el resultado
df_filtered.head()
```

```python
# Describir la columna 'Duration' para obtener una vista general de los valores
print(df_filtered['Duration'].describe())

# Calcular los cuartiles y el IQR (rango intercuartílico)
Q1 = df_filtered['Duration'].quantile(0.25)
Q3 = df_filtered['Duration'].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites inferior y superior para identificar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los outliers
df_no_outliers = df_filtered[(df_filtered['Duration'] >= lower_bound) & (df_filtered['Duration'] <= upper_bound)]

# Mostrar la cantidad de reuniones antes y después de eliminar outliers
print(f"Cantidad de reuniones antes de eliminar outliers: {len(df_filtered)}")
print(f"Cantidad de reuniones después de eliminar outliers: {len(df_no_outliers)}")

# Verificar el resultado
df_no_outliers.head()
```

```python
# Eliminar reuniones cuya duración sea mayor a un umbral (por ejemplo, 500 minutos)
df_filtered = df_filtered[df_filtered['Duration'] <= 500]
```

```python
# Exportar el DataFrame filtrado a un archivo CSV para Tableau
df_filtered.to_csv('reuniones_filtradas.csv', index=False, encoding='utf-8')

# Confirmación
print("El archivo 'reuniones_filtradas.csv' ha sido exportado correctamente.")
```
